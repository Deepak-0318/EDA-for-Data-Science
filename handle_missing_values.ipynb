Handling Missing Values

1. Deleting the columns with missing data

```python
updated_df = df.dropna(axis=1)  # Dropping columns with missing values
```

```python
updated_df.info()
```

2. Deleting the rows with missing data

```python
updated_df = df.dropna(axis=0)
```

```python
updated_df.info()
```

3. Filling the missing values - Imputation

```python
df['Age'].mean()
```

```python
df['Age'].median()
```

```python
#fillna: fills the null records
#dropna: drops the null records
updated_df = df
updated_df['Age']=updated_df['Age'].fillna(df['Age'].mean())
updated_df.info()
```

```python
updated_df1 = df
updated_df1['Age']=updated_df1['Age'].fillna(df['Age'].median())
updated_df1.info()
```

4. Forward and Backwrd Filling - Imputation

```python
df1 = df
```

```python
#bfill: backward fill
df1['Age'] = df1['Age'].bfill()
df1.info()
```
