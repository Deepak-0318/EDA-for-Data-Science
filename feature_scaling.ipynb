Normalization

```python
df.head()
```

```python
new_df = pd.DataFrame(df,  columns=['Age', 'Tenure'])
new_df.head()
```

```python
new_df['Age']=new_df["Age"].fillna(new_df['Age'].mean())
```

```python
scaler = MinMaxScaler() #Instantiating the min max scaler
normalized_df = scaler.fit_transform(new_df)
print(normalized_df)
```

Another Example

```python
x_array = np.array([[2],[3],[5],[7],[6]])

scaler = MinMaxScaler()  # instantiate the scaler
normalized_df = scaler.fit_transform(x_array)
print(normalized_df)
```
